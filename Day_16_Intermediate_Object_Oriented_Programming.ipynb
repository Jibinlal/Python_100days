# **Object Oriented Programming**

- Until now we have seen procedural programming- -which gets complicated when the program gets long. The alternative to this approach is Object oriented programming.
- Object oriented programming splits a complex projects into Separate small modules which can be easy to handle and are re-usable
- It is called 'Object Oriented as we are trying to model a real world object'

Accessing the attributes of objects:

```python
# What to do with the object - OBJECT ATTRIBUTES

# Import Screen class as well 
from turtle import Turtle, Screen
timmy = Turtle()
print (timmy)

# creating an object 
my_screen = Screen()
# acessing the attribute of the object
print(my_screen.canvheight)
```

OBJECT METHODS:

```python
# Object Methods

# Import Screen class as well 
from turtle import Turtle, Screen
timmy = Turtle()
print (timmy)

# creating an object 
my_screen = Screen()
# acessing the attribute of the object
print(my_screen.canvheight)

# Acessing a method to make the item as turtle
timmy.shape("turtle")
# accessing the method of the object. Access the method to make it not to disappear the window
my_screen.exitonclick()
```

Output of above (from Pycharm):
The window doesn't disappear now. Only disappears after a click on the window. Also the shape of turtle appear

```python
# Modifying Further

from turtle import Turtle, Screen
timmy = Turtle()
print (timmy)

my_screen = Screen()
print(my_screen.canvheight)
timmy.shape("turtle")
# change color
timmy.color("red","green")
# move 100 paces
timmy.forward(100)

my_screen.exitonclick()
```

```python
from prettytable import PrettyTable
# Documentation - https://code.google.com/archive/p/prettytable/wikis/Tutorial.wiki

x = PrettyTable()
x.add_column("City name",["Adelaide","Brisbane","Darwin","Hobart","Sydney","Melbourne","Perth"])
x.add_column("Area", [1295, 5905, 112, 1357, 2058, 1566, 5386])
x.add_column("Population", [1158259, 1857594, 120900, 205556, 4336374, 3806092, 1554769])
x.add_column("Annual Rainfall",[600.5, 1146.4, 1714.7, 619.5, 1214.8, 646.9, 869.4])

# changing the align variable value to modify the alignment of output
x.align = "l"
print(x)
```

Would you explain, in simple terms, exactly what object-oriented software is? -- By Steve Jobs

Here’s an example: If I’m your laundry object, you can give me your dirty clothes and send me a message that says, “Can you get my clothes laundered, please.” I happen to know where the best laundry place in San Francisco is. And I speak English, and I have dollars in my pockets. So I go out and hail a taxicab and tell the driver to take me to this place in San Francisco. I go get your clothes laundered, I jump back in the cab, I get back here. I give you your clean clothes and say, “Here are your clean clothes.”

You have no idea how I did that. You have no knowledge of the laundry place. Maybe you speak French, and you can’t even hail a taxi. You can’t pay for one, you don’t have dollars in your pocket. Yet I knew how to do all of that. And you didn’t have to know any of it. All that complexity was hidden inside of me, and we were able to interact at a very high level of abstraction. That’s what objects are. They encapsulate complexity, and the interfaces to that complexity are high level.
