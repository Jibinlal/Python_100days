When to use while and for:


1.   FOR is used when you need to iterate over a range or set of items ; WHILE is used when you need to check a condition and the execute a set of operations
2.   FOR loops have a range or set of items; WHILE loops runs until the condition is satisfied - which can create an infinte loop
3. While loops are best when you are not sure how much time to run the loop and when you have a condition to satisfy
