CREATING CLASSESS SYNTAX

```python
# creating a class; the name should be in PascalCase
class UserClass:
  # Pass is used when you want an empty function/Class. If you don't use Pass, it would throw an error
  pass
  
# creating a object from class
user_1 = UserClass()

# creating attribute from class; just use the object and assign any random variable you want
user_1.username = "Angela"
user_1.id = "001"

print(user_1.username)
```

In above code where we assign attribute values, we just assigned the attribute values to dervied object. Is there way to pre-initialize the attributes when you create a object from Class? - Yes, using Constructor!

We can use a special function (init function) to initialize attributes

image.png

```python

class UserClass:
  # Everytime a object is created from this class, the init function will be run
  def __init__(self):
    print("new user being created")
  

user_1 = UserClass()
```

**Setting Attributes in constructor:**

image.png

This is similar to assigning the value of attribute directly

image.png

```python
class UserClass:
  def __init__(self,user_id):
    # user id passed by the user is assigned to variable called ID
    self.id = user_id
  
# we are passing the user id value to class
user_1 = UserClass("001")

# As you can see, the id variable is created and assigned by default
print(user_1.id)
```

```python
# Adding more variables

class UserClass:
  def __init__(self,user_id,username):
    self.id = user_id
    self.username = username
    # YOU CAN ALSO ASSIGN A DEFAULT VALUE
    self.followers = 0
  
user_1 = UserClass("001","Angela")

print(user_1.id)
print(user_1.username)
print(user_1.followers)
```

**ADDING METHODS TO A CLASS**

Similar to initializing a function

image.png

```python
# Adding more variables

class UserClass:
  def __init__(self,user_id,username):
    self.id = user_id
    self.username = username
    # YOU CAN ALSO ASSIGN A DEFAULT VALUE
    self.followers = 0
    self.following = 0
  
  # SELF is a imp keyword. It denotes the object that is being created from the Class!
  def follow(self,user):
    user.followers +=1
    self.following +=1
  
user_1 = UserClass("001","Angela")
user_2 = UserClass("002","Jibin")

user_1.follow(user_2)

print(user_1.followers)
print(user_1.following)
print(user_2.followers)
print(user_2.following)
```
